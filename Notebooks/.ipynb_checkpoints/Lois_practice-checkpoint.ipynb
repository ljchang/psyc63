```python
# change to 0,1
df.loc[df['romantic']=='yes','romantic']=1
df.loc[df['romantic']=='no','romantic']=0
# change to numeric
df.romantic = pd.to_numeric(df.romantic)

logm = smf.logit('romantic ~ studytime ',data=df).fit()
print logm.summary()
```
